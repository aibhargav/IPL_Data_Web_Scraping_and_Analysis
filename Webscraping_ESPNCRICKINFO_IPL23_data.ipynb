```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Function to scrape the main webpage and find links containing "full-scorecard"
def scrape_main_page(url):
    # Send a GET request to the main webpage
    response = requests.get(url)
    
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all the links on the main page
    links = soup.find_all('a')
    
    # Loop through the links and store the URLs containing "full-scorecard"
    full_scorecard_links = []
    for link in links:
        # Get the href attribute of the link
        href = link.get('href')
        
        # Check if the link contains "full-scorecard" and "https://www.espncricinfo.com/"
        if href and 'full-scorecard' in href:
            full_scorecard_links.append('https://www.espncricinfo.com/' + href)
    
    # Process the full-scorecard links and store the results
    data_list = []
    for i, link in enumerate(full_scorecard_links):
        # Send a GET request to the full-scorecard link
        response = requests.get(link)
        
        # Get the HTML content
        html_content = response.text
        
        # Read the HTML tables into a list of DataFrames
        tables = pd.read_html(html_content)
        
        # Append the DataFrame to the data_list
        data_list.append(tables)
        
        # Store the DataFrame in a separate variable (data1, data2, ...)
        globals()["data{}".format(i+1)] = tables
    
    # Print the stored DataFrames (optional)
    for i, data in enumerate(data_list):
        print(f"data{i+1}:")
        print(data)

# Main function
def main():
    # URL of the main webpage
    main_url = 'https://www.espncricinfo.com/series/indian-premier-league-2023-1345038/match-schedule-fixtures-and-results'
    
    # Call the function to scrape the main webpage and process the full-scorecard links
    scrape_main_page(main_url)

# Run the main function
if __name__ == '__main__':
    main()
```
